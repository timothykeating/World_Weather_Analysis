```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests

# Import the API key.
from config import g_key
```

```python
city_data_df = pd.read_csv("weather_data/cities.csv")
city_data_df.head()
```

```python
city_data_df.dtypes
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
lowtemp = city_data_df.loc[city_data_df['Max Temp']<=0]
lowtemp
```

```python
# replace negative max temps with 0
max_temp=city_data_df['Max Temp']

temps=[]

for temp in max_temp:
    temps.append(max(temp,0))
temps


### In the for loop, we're using the max() function to get the largest value between the temp and 0. 
### If the temp is less than 0, then 0 will be added to the list in its place. 
### Otherwise, the temp is added to the list.
```

```python
# 1. Assign the locations to an array of latitude and longitude pairs.
locations = city_data_df[['Lat','Lng']]


# 2. Assign the weights variable to some values.
### we assigned the weights variable as temps one cell above
### you cant have negative weights in this method, so using city_data_df['Max Temp'] will not work


# 3. Assign the figure variable to the gmaps.figure() attribute.
### use center= to center the map, 
### use zoom to zoom in to where you only see how much you want
fig = gmaps.figure(center=(30.0,31.0),zoom_level=1.85)


# 4. Assign the heatmap_layer variable to the heatmap_layer attribute and add in the locations.
### use temps, lowercase, new object created above, for the weights
### dissipating, max_intensity and point_radius are all params that help show the colors better on the heatmap
heatmap_layer = gmaps.heatmap_layer(locations, weights=temps, dissipating=False, max_intensity=300, point_radius=4)


# 5. Add the heatmap layer.
fig.add_layer(heatmap_layer)


# 6. Call the figure to plot the data.
fig
```

```python
# Get Travel Destinations

# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = tempfilter_df[["City", "Country", "Max Temp", "Lat", "Lng"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head()
```

```python
# Retrieve Hotels from a Nearby Search

# # Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
```

```python
## We can use the iterrows() function to perform the iteration; 
## however, we need to provide the index and the row in the for loop using this syntax: 
## for index, row in df.iterrows().
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"
# params


    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
hotels
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    
    # Grab the first hotel from the results and store the name.
    try:
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
        print("All set")
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df
```

```python
###########################
## Create a Maximum Temperature Heatmap from a Hotel DataFrame

# Add a heatmap of temperature for the vacation spots to hotel_df this time
locations = hotel_df[["Lat", "Lng"]]
max_temp = hotel_df["Max Temp"]

fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)

heat_layer = gmaps.heatmap_layer(locations, weights=max_temp, dissipating=False,
             max_intensity=300, point_radius=4)

## marker_layer will be markers added to indicate hotel l;ocations
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

fig.add_layer(heat_layer)

## add the marker layer 
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
